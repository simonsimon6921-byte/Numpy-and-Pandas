### Numpy Array Operations :

#### 1. Create a 1D Numpy Array :

```python
import numpy as np


temperatures_w1 = np.array([22.5,25.3,20.8,23.4,26.1,24.8,21.9])
temperatures_w1
```

```python
print(temperatures_w1)
```

#### 2. Inspection and Properties :

```python
temperatures_w1 = np.array([22.5,25.3,20.8,23.4,26.1,24.8,21.9],dtype=int)
temperatures_w1
```

#### 3. Array Operations :

```python
Fahrenheit = (temperatures_w1 * 9/5)+32
print(Fahrenheit)
```

```python
print("Maximum temperature:", np.max (temperatures_w1))
print("Minimum temperature:", np.min (temperatures_w1))
print("mean temperature:", np.mean (temperatures_w1))
```

#### 4. Array Slicing and Indexing :

```python
first_three = temperatures_w1[:3]
print("first three days:")
print(first_three)
```

```python
last_two = temperatures_w1[-2]
print("last two days:")
print(last_two)
```

```python
middle_three = temperatures_w1[2:5]
print("middle three days:")
print(middle_three)
```

#### 1. Creating Pandas DataFrame:

```python
import pandas as pd
mark = pd.Series([95, 92, 89, 85, 80],index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5'])
print(mark)
```

#### 2. Indexing and Slicing:

```python
print("\nMark of 1st rank student:")
print(mark['Rank1'])
```

```python
print("\n top 3 ranks using loc:")
print(mark.loc[:'Rank3'])
```

```python
print("\n mark of the 3rd-ranked student using iloc:")
print(mark.iloc[2])
```

```python
print("\n ranks where the marks are greater then 90:")
print(mark[mark >90])
```

#### 3. Manipulating Series:

```python
mark ['Rank1'] = 100
print("\n Modify the mark of the student with the 1st rank to 100 :")
print(mark)
```

```python
mark = mark.drop('Rank5')
```

```python
print (mark)
```

```python
cgpa = mark / 10

print("\nCGPA:")
print(cgpa)
```

### Pandas DataFrame:

#### 1. Creating Pandas DataFrame:

```python
import pandas as pd
transactions = pd.DataFrame({
    'TransactionID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'ProductCategory': [
        'Electronics', 'Clothing', 'Electronics', 'Furniture',
        'Clothing', 'Electronics', 'Furniture',
        'Clothing', 'Furniture', 'Electronics'
    ],
    'Region': [
        'North', 'South', 'North', 'East', 'West',
        'North', 'East', 'West', 'South', 'North'
    ],
    'Amount': [200, 150, 300, 450, 200, 250, 300, 180, 350, 400]
})
print("\n transactions DataFrame:")
print(transactions)
```

#### 2. Data Exploration:

```python
print("/n Head of DataFrame:")
print(transactions.head())
```

```python
print("\nTail of DataFrame:")
print(transactions.tail())
```

```python
print("\n shape of DataFrame:")
print(transactions.shape)
```

```python
print("\n column names of DataFrame:")
print(transactions.columns)
```

```python
print("\n data type of DataFrame:")
print(transactions.dtypes)
```

```python
print("\n ProductCategory and Amouunt columns")
print(transactions[['ProductCategory','Amount']])
```

```python
print("\n last 3 columns:")
print(transactions.iloc[:,-3:])
```

```python
print("\n rows where the Region is North and Amount > 200:")
filtered_data =transactions [
    (transactions['Region'] == 'North') &
    (transactions['Amount'] > 200)
]
print(filtered_data)
```

```python
print("\nValue counts for ProductCategory:")
print(transactions['ProductCategory'].value_counts())
```

```python
print("\nUnique Regions:")
print(transactions['Region'].unique())
```

```python
print("\nMean Amount by Region:")
print(transactions.groupby('Region')['Amount'].mean())
```

#### 3. Manipulating the DataFrame:

```python
transactions.loc [transactions['TransactionID'] == 102, 'Amount'] =165
print("\n Modify the Amount for the transactionID102:")
print(transactions)
```

```python
transactions ['Discount'] = transactions ['Amount'] * 0.10
print("\n After adding Discount column:")
print(transactions)
```

```python
transactions = transactions[transactions['TransactionID'] != 109]
print("\nAfter removing TransactionID 109:")
print(transactions)
```

```python
transactions.drop('Discount', axis=1, inplace=True)
print("\nAfter deleting Discount column:")
print(transactions)
```
